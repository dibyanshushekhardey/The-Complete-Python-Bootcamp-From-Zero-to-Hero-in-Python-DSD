# lambda expressions map and filter

```python
def square(num):
    return num ** 2
```

```python
my_nums = [1, 2, 3, 4, 5]
```

```python
map(square, my_nums)
```

```python
for item in map(square, my_nums):
    print(item)
```

```python
list(map(square, my_nums))
```

```python
def splicer(mystring):
    if len(mystring)%2 == 0:
        return 'EVEN'
    else:
        return mystring[0]
```

```python
names = ['Andy', 'Eve', 'Sally']
```

```python
list(map(splicer, names))
```

```python
def check_even(num):
    return num % 2 == 0
```

```python
mynums=[1, 2, 3, 4, 5, 6]
```

```python
list(filter(check_even, mynums))
```

```python
for n in filter(check_even, mynums):
    print(n)
```

```python
# Global
name = 'THIS IS A GLOBAL STRING'

def greet():
    # Enclosing
    name = 'Sammy'
    
    def hello():
        # local
        name = 'I am a local'
        print('Hello ' + name)
        
    hello()

```

```python
greet()
```

```python
help(len)
```

```python
x = 50

def func():
    global x
    print(f'X is {x}')
    
    # local reassignment on a global variable!
    x = 'NEW VALUE'
    print(f'I just locally changed global x to {x}')
```

```python
print(x)
```

```python
func()
```

```python
print(x)
```

```python
func(x)
```

```python
print(x)
```

```python
x = 50
def func(x):
    #global x
    print(f'X is {x}')
    
    x = 'New value'
    print(f'I just locally changed global X to {x}')
    return x
```
