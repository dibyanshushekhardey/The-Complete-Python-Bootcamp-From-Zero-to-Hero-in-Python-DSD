####  makes twenty

```python
def makes_twenty(n1, n2):
    if n1 + n2 == 20:
        return True
    elif n1 ==  20:
        return True
    elif n2 == 20:
        return True
    else:
        return False
```

#### old macdonald

```python
def old_macdonald(name):
    first_letter = name[0]
    inbetween = name[1:3]
    fourth_letter = name[3]
    rest = name[4:]
    
    return first_letter.upper() + inbetween + fourth_letter.upper() + rest
```

```python
old_macdonald('macdonald')
```

```python
def old_macdonald(name):
    first_half = name[:3]
    second_half = name[3:]
    
    return first_half.capitalize() + second_half.capitalize()
```

```python
old_macdonald('macdonald')
```

#### Master Yoda

```python
def master_yoda(text):
    wordlist = text.split()
    reversed_word_list = wordlist[::-1]
    return ' '.join(reversed_word_list)
```

```python
master_yoda('I am home')
```

```python
master_yoda('We are ready')
```

#### find 33

```python
def has_33(nums):
    for i in range(0, len(nums)-1):
        if nums[i] == 3 and nums[i+1] == 3:
            return True
    return False
```

```python
has_33([1, 3, 3])
```

```python
has_33([1, 3, 2, 3])
```

#### paper doll

```python
def paper_doll(text):
    result = ''
    
    for char in text:
        result += char*3
    return result
```

#### blackjack

```python
def blackjack(a,b,c):
    if sum([a, b, c]) <= 21:
        return sum([a, b, c])
    elif 11 in [a, b, c] and sum([a, b, c]) <= 31:
        return sum([a, b, c]) - 10
    else:
        return 'BUST'
```

```python
blackjack(5, 6, 7)
```

```python
blackjack(9, 9, 9)
```

```python
blackjack(9,9,11)
```

#### Summer of '69

```python
def summer_69(arr):
    total = 0
    add = True
    
    for num in arr:
        while add:
            if num!= 6:
                total += num
                break
            else:
                add = False
        while not add:
            if num != 9:
                break
            else:
                add = True
                break
    return total
```

```python
summer_69([1, 3, 5])
```

```python
summer_69([4, 5, 6, 7, 8, 9])
```

```python
summer_69([2, 1, 6, 9, 11])
```

#### spy games

```python
def spy_games(nums):
    code=[0, 0, 7, 'x']
    # [0, 0, 7, 'x']
    # [0, 7, 'x']
    
    for num in nums:
        if num == code[0]:
            code.pop(0)
            
    return len(code) == 1
            
```

```python
spy_games([1,2,4,0,0,7,5])
```

```python
spy_games([1, 0, 2, 4, 0, 5, 7])
```

```python
spy_games([1,7,2,0,4,5,0])
```

#### count primes

```python
def count_primes(num):
    # check for 0 or 1 input
    if num < 2:
        return 0
    # 2 or greater
    # store our prime numbers
    primes = [2]
    # counter going up to input num
    x = 3
    
    # x is going thorugh every number upto input num
    while x <= num:
        # check if x is prime
        for y in range(3, x, 2):
            if x % y == 0:
                x += 2
                break
        else:    # for else statement 
            primes.append(x)
            x += 2
    print(primes)
    return len(primes)
```

```python
count_primes(100)
```
