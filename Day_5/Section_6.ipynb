```python
work_hours = [('Abby', 100), ('Billy', 4000), ('Cassie', 800)]
```

```python
def employee_check(work_hours):
    current_max = 0
    employee_of_month = ''
    
    for employee, hours in work_hours:
        if hours > current_max:
            current_max = hours
            employee_of_month = employee
        else:
            pass
    # return
    return(employee_of_month, current_max)
```

```python
result = employee_check(work_hours)
```

```python
name, hours = employee_check(work_hours)
```

```python
name
```

```python
hours
```

```python
name, hours, location = employee_check(work_hours)
```

```python
def player_guess():
    
    guess=''  # placeholder for the user guess
    while guess not in ['0', '1', '2']:
        guess = input("Pick a number: 0, 1, or 2: ")
    
    return int(guess)
```

```python
player_guess()
```

```python
my_index = player_guess()
```

```python
my_index
```

```python
def check_guess(mylist, guess):
    if mylist[guess] == '0':
        print('Correct!')
    else:
        print('Wrong guess!')
        print(mylist)
```

```python
# initial list
mylist = [' ', 'O', ' ']
# shuffle list
mixedup_list = shuffle_list(mylist)
# user guess
guess = player_guess()
# check guess
check_guess(mixedup_list, guess)
```

```python
def myfunc(**kwargs):
    print(kwargs)
    if 'fruit' in kwargs:
        print('My fruit of choice is {}'.format(kwargs['fruit']))
    else:
        print('I did not fin any fruit here')
```

```python
def myfunc(x):
    l = []
    for i in x:
        if x.index(i) % 2 == 0:
            i = i.upper()
            l.append(i)
        else:
            i = i.lower()
            l.append(i)
    return ''.join(l)
```
