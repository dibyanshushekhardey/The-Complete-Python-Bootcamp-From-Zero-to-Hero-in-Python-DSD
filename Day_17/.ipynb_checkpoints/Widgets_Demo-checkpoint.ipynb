```python
from ipywidgets import interact, interactive, fixed
import ipywidgets as widgets
```

```python
def func(x):
    return x
```

```python
interact(func, x='Hello')
```

```python
@interact(x=True, y=1.0)
def g(x, y):
    return (x, y)
```

```python
# fixed
def h(p, q):
    return(p, q)
```

```python
interact(h, p=5, q=fixed(20))
```

```python
# widget abbreviations
interact(func, x=widgets.IntSlider(min=-10,max=30, step=1, value=10));
```

```python
# step size
interact(func, x=(0, 8, 2))
```

```python
interact(func, x=(0.0, 10.0));
```

```python
interact(func, x=(0.0,10.0,0.01));
```

```python
@interact(x=(0.0, 20.0, 0.5))
def h(x=5.5):
    return x
```

```python
interact(func, x=['apples', 'oranges']);
```

```python
interact(func, x={'one':10, 'two':20});
```

```python
# function annotations with interact
```

```python
def f(x:True):
    return x
```

```python
interact(f);
```

```python
from IPython.display import display

def f(a, b):
    display(a+b)
    return a + b
```

```python
w = interactive(f, a=10, b = 20);
```

```python
# linking two widgets
a = widgets.FloatText()
b = widgets.FloatSlider()
display(a, b)

mylink = widgets.jslink((a, 'value'), (b, 'value'))
```

```python
# unlinking widgets
mylink.unlink()
```
