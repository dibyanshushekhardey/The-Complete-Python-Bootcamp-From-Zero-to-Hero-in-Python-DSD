```python
mystring = 'Sammy'

```

```python
for letter in mystring:
    if letter == 'a':
        continue  # goes to the top of the closest enclosing loop
    print(letter)
```

```python
for letter in mystring:
    if letter == 'a':
        break  # breaks out of the current closest enclosing loop
    print(letter)
```

```python
x = 0

while x < 5:
    if x == 2:
        break
    print(x)
    x += 1
    
```

## List comprehensions

```python
mystring = 'hello'
```

```python
mylist = []

for letter in mystring:
    mylist.append(letter)
```

```python
mylist
```

```python
mylist = [letter for letter in mystring]
```

```python
mylist
```

```python
mylist = [two for two in 'word'] 
```

```python
mylist
```

```python
mylist = [num**2 for num in range(0, 11)]
```

```python
mylist
```

```python
mylist = [x**2 for x in range(0, 11) if x % 2 ==0]
```

```python
mylist
```

```python
celsius = [0, 10, 20, 34.5]

fahrenheit = [((9/5)*temp + 32) for temp in celsius]
```

```python
fahrenheit
```

```python
fahrenheit = []
for temp in celsius:
    fahrenheit.append(((9/5)*temp + 32))
```

```python
fahrenheit
```
