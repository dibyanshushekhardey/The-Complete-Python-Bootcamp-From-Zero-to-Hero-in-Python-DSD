```python
def hello(name='Dibyanshu'):
    print("The hello() function has been executed!")
    
    def greet():
        return "\t This is the greet() func inside hello!"
    
    def welcome():
        return "\t This is welcome() inside hello!"
    
    print(greet()) ## calling greet function
    print(welcome())
    print("This is the end of the hello function!")
```

```python
hello()
```

```python
welcome() # calling from outside the hello function , limited scope
```

```python
def hello(name='Dibyanshu'):
    print("The hello() function has been executed!")
    
    def greet():
        return "\t This is the greet() func inside hello!"
    
    def welcome():
        return "\t This is welcome() inside hello!"
    
    print("I am going to return a function!")
    
    if name == 'Dibyanshu':
        return greet
    else:
        return welcome
```

```python
my_new_func = hello('Dibyanshu')
```

```python
my_new_func
```

```python
print(my_new_func())
```

```python
def new_decorator(original_func):
    
    def wrap_func():
        print('Some extra code, before the original function')
        
        original_func()
        
        print('Some extea code, after the original function')
        
    return wrap_func
```

```python
def func_needs_decorator():
    print("I want to be decorated!!!")
```

```python
func_needs_decorator()
```

```python
decorated_func = new_decorator(func_needs_decorator)
```

```python
decorated_func()
```

```python
# special syntax
@new_decorator
def func_needs_decorator():
    print("I want to be decorated!!!")
```

```python
func_needs_decorator()
```
