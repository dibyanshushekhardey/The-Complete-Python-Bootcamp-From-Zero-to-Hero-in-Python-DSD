```python
import random

suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 
          'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 
          'Ten':10, 'Jack':10, 'Queen':10, 'King':10, 'Ace':11}

playing= True
```

```python
class Card:
    def __init__(self, suit, rank):
        self.suit = suit
        self.rank = rank
    def __str__(self):
        return self.rank + " of " + self.suit
    
```

```python
# deck class

class Deck:
    def __init__(self):
        self.deck = []
        for suit in suits:
            for rank in ranks:
                self.deck.append(Card(suit, rank))
    def __str__(self):
        deck_comp =''
        for card in self.deck:
            deck_comp += '\n' + card.__str__()
            
        return "The deck has: " + deck_comp
    
    def shuffle(self):
        random.shuffle(self.deck)
    
    def deal(self):
        single_card = self.deck.pop()
        return single_card
```

```python
class Hand:
    def __init__(self):
        self.cards = []
        self.value = 0
        self.aces = 0
        
    def add_card(self, card):
        # caard passed in
        # from deck.deal() --> single card(suit, rank)
        self.cards.append(card)
        self.value += values[card.rank]
        
        # track aces
        if card.rank == 'Ace':
            self.aces += 1
            
    
    def adjust_for_ace(self):
        # if totak_value < 21 and i still have ace
        # then change my ace to be a 1 instead of an 11
        
        while self.value > 21 and self.aces:
            self.value -= 10
            self.aces -= 1
            
```

```python
class Chips:
    def __init__(self):
        self.total = 100
        self.bet = 0
    def win_bet(self):
        self.total += self.bet
    def lose_bet(self):
        self.total -= self.bet
```

```python
# function definitions
def take_bet(chips):
    while True:
        try:
            chips.bet = int(input("How many chips would you like to bet? "))
        except:
            print("Sorry please provide an integer")
        else:
            if chips.bet > chips.total:
                print("Sorry, you do not have enough chips! you have: {}".format(chips.total))
            else:
                break
```

```python
# write a function for taking hits
def hit(deck, hand):
    single_card = deck.deal()
    hand.add_card(single_card)
    hand.adjust_for_ace()
    
```

```python
# function for prompting the player to hit or stand
def hit_or_stand(deck, hand):
    global playing 
    while True:
        x = input("Hit or Stand? Enter h or s") #HIT hh stand
        
        if x[0].lower() == 'h':
            hit(deck, hand)
        elif x[0].lower() == 's':
            print("Player stands! Dealer's turn")
            playing = False
        else:
            print("Sorry I did not understand that, please enter h or s only!")
            continue
        break
```

```python
# function to display cards
def show_some(player, dealer):
    # show ony one of the dealer's cards
    print("\n Dealer's Hand: ")
    print("First card hidden!")
    print(dealer.cards[1])
    
    # show all (2 cards) of the player's hand/cards
    print("\n Player's hand:")
    for card in player.cards:
        print(card)
    
def show_all(player, dealer):
    # show all dealer's cards
    print("\n Dealer's hand:")
    for card in dealer.cards:
        print(card)
    print("\n Dealer's hand: ", *dealer.cards)
    
    # Calculate and display the value(j+k = 20)
    print(f"Value of Dealer's hand is: {dealer.value}")
    
    # show all the player's cards
    print("\n Player's hand:")
    for card in player.cards:
        print(card)
    print(f"Value of Player's hand is: {player.value}")
```
