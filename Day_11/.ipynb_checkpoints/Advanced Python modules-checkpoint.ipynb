```python
def func_one(n):
    return [str(num) for num in range(n)]
```

```python
func_one(10)
```

```python
def func_two(n):
    return list(map(str, range(n)))
```

```python
func_two(10)
```

```python
import time
```

```python
# current time before
start_time = time.time()
# run code
result = func_two(1000000)
# cureent time after running code
end_time = time.time()
# elapsed time
elapsed_time = end_time - start_time

print(elapsed_time)
```

```python
# current time before
start_time = time.time()
# run code
result = func_two(10)
# cureent time after running code
end_time = time.time()
# elapsed time
elapsed_time = end_time - start_time

print(elapsed_time)
```

```python
import timeit
```

```python
stmt = '''
func_one(100)
'''
```

```python
setup = '''
def func_one(n):
    return [str(num) for num in range(n)]
'''
```

```python
timeit.timeit(stmt, setup, number=1000000)
```

```python
stmt2 = '''
func_two(100)
'''
```

```python
setup2 = '''
def func_two(n):
    return list(map(str, range(n)))
'''
```

```python
timeit.timeit(stmt2, setup2, number=1000000)
```

```python
%%timeit
func_one(100)
```

```python
%%timeit
func_two(100)
```

```python
f = open('fileone.txt', 'w+')
f.write('ONE FILE')
f.close()
```

```python
f = open('filetwo.txt', 'w+')
f.write('TWO FILE')
f.close()
```

```python
import zipfile
```

```python
comp_file = zipfile.ZipFile('comp_file.zip', 'w')
```

```python
comp_file.write('fileone.txt', compress_type=zipfile.ZIP_DEFLATED)
```

```python
comp_file.write('filetwo.txt', compress_type=zipfile.ZIP_DEFLATED)
```

```python
comp_file.close()
```

```python
zip_obj = zipfile.ZipFile('comp_file.zip', 'r')
```

```python
zip_obj.extractall('extracted_content')
```
