```python
from collections import Counter
```

```python
mylist = [1, 1, 1, 1, 1, 2, 3, 3, 3, 2, 2, 2, 3, 3]
```

```python
Counter(mylist)
```

```python
mylist = ['a', 'a', 10, 10, 10]
```

```python
Counter(mylist)
```

```python
Counter('adhskjfgksgfkkjsdgfjsfk')
```

```python
sentence = "How many times does each word show up in this sentence"
```

```python
Counter(sentence.lower().split())
```

```python
letters = 'gdfhjagfdkfshfghjsdfjkejf'
```

```python
c = Counter(letters)
```

```python
c
```

```python
c.most_common(3)
```

```python
c
```

```python
list(c)
```
