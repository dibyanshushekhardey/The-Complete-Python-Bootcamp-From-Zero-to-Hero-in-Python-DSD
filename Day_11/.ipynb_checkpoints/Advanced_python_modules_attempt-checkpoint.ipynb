```python
def search(file,pattern= regex_pattern):
    f = open(file,'r')
    text = f.read()
    
    if re.search(pattern,text):
        return re.search(pattern,text)
    else:
        return ''
```

```python
folder_path = '/Users/dibyanshu/Downloads/The-Complete-Python-Bootcamp-From-Zero-to-Hero-in-Python-DSD/Day_11/extracted_content'
```

```python
results = []
for folder , sub_folders , files in os.walk(folder_path): 
    for f in files:
        full_path = folder+'/'+f # Replace \\ by / if using a Unix based OS
        results.append(search(full_path)) 
```

```python
for r in results:
    if r != '':
        print(r.group())
```
