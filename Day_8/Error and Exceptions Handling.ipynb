## try, except, finally 

```python
try:
    f = open('testfile', 'r')
    f.write("Write a test line")
except TypeError:
    print("There was a type error!")
except:
    print('All other exceptions')
finally:
    print("I always run!")
```

```python
def ask_for_int():
    while True:
        try:
            result = int(input("Please provide number: "))
        except:
            print("whoops! that is not a number!")
            continue
        else: # if there is not an error
            print("Yes thank you")
            break # breaks the loop then gies to finally
        
```

```python
ask_for_int()
```
